```python
import csv
from googletrans import Translator
```

```python
translator = Translator()
```

```python
# Read file
filename = "english_words.csv" # file name
with open(filename) as f:
    reader = csv.reader(f)
    next(reader) # skip the header row
    english_words = [row[0] for row in reader]
```

```python
translated_words = []
for word in english_words:
  try:
    translation = translator.translate(word, dest='km')
    translated_words.append(translation.text)
  except:
    translated_words.append('') # if google can't translate then translated word is empty
    # next(reader) # if google can't translate then you skip loop
```

```python
  output_filename = "translated_words.csv"
  with open(output_filename, 'w', newline='') as f:
      writer = csv.writer(f)
      writer.writerow(["English Words", "Translated Words"]) # write header row
      for i in range(len(english_words)):
          writer.writerow([english_words[i], translated_words[i]])
```
